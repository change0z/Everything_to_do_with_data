```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    try:
        container = soup.find('div', class_='item-cell')  # Adjust the class name
        titles = container.find_all('a', class_='item-title')
        for title in titles:
            processor_name = title.get_text(strip=True)
            print(f"Processor Name: {processor_name}")
    except AttributeError as e:
        print(f"Error: {e}")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    titles = soup.find_all('a', class_='item-title')  # Use find_all to get all matching elements
    for i, title in enumerate(titles, start=1):
        processor_name = title.get_text(strip=True)
        print(f"{i}. Processor Name: {processor_name}")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
