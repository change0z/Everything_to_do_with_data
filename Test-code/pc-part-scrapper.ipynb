```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python

url = 'https://www.newegg.com/global/uk-en/p/pl?d=cpu&N=601393085%20101582716'
response = requests.get(url)
print(response.status_code)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    price_elements = soup.find_all('li', class_='price-current')  # Find all price elements
    for i, price_element in enumerate(price_elements, start=1):
        currency = price_element.contents[1]  # Get the currency symbol
        whole_part = price_element.strong.text  # Get the whole number part
        decimal_part = price_element.sup.text  # Get the decimal part
        full_price = f"{currency}{whole_part}{decimal_part}"
        print(f"{i}. Price: {full_price}")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import re

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    titles = soup.find_all('a', class_='item-title')  # Use find_all to get all matching elements
    for i, title in enumerate(titles, start=1):
        full_text = title.get_text(strip=True)
        
        # Extract processor name (assuming it always starts with "Intel Core")
        processor_name = re.search(r'(Intel Core [^\s-]+(?:-[^\s]+)?)', full_text)
        
        if processor_name:
            print(f"{i}. {processor_name.group(1)}")
        else:
            print(f"{i}. Unable to parse: {full_text[:50]}...")  # Print first 50 characters of the full text
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
